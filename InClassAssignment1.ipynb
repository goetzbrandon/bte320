```python
sales = float(input("Enter sales ($): "))
tax = float(input("Enter tax rate (%): "))
print(f"Sales: ${sales}")
print(f"Cost of Goods Sold: ${sales*.4}")
print(f"Gross Profit: ${sales-sales*.4}")
print(f"Taxes: ${(sales-sales*.4)*(tax/100)}")
print(f"Net Income: ${(sales-sales*.4)-((sales-sales*.4)*(tax/100))}")
```

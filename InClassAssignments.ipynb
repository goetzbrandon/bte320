```python
sales = float(input("Enter sales ($): "))
tax = float(input("Enter tax rate (%): "))
print(f"Sales: ${sales}")
print(f"Cost of Goods Sold: ${sales*.4}")
print(f"Gross Profit: ${sales-sales*.4}")
print(f"Taxes: ${(sales-sales*.4)*(tax/100)}")
print(f"Net Income: ${(sales-sales*.4)-((sales-sales*.4)*(tax/100))}")
```

```python
x = float(input("Give the first value: "))
operator = str(input("Provide the operator: "))
y = float(input("Give the second value: "))
if operator == "+":
  z = x + y
elif operator == "-":
  z = x - y
elif operator == "*":
  z = x * y
elif operator == "/":
  z = x / y
else:
  print("Not a valid operator.")


print(z)
```

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
op = input("Choose numerical operator (e.g., +, -, *, etc.): ")

res = None

if op == "+":
  res = num1 + num2
elif op == "-":
  res = num1 - num2
elif op == "*":
  res = num1 * num2
elif op == "/":
  res = num1 / num2
elif op == "//":
  res = num1 // num2
elif op == "**":
  res = num1 ** num2
elif op == "%":
  res = num1 % num2
else:
  print("Invalid operation")

print(f"Result: {res}")
```

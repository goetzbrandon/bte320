```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
op = input("Choose numerical operator (e.g., +, -, *, etc.): ")

res = None

if op == "+":
  res = num1 + num2
elif op == "-":
  res = num1 - num2
elif op == "*":
  res = num1 * num2
elif op == "/":
  res = num1 / num2
elif op == "//":
  res = num1 // num2
elif op == "**":
  res = num1 ** num2
elif op == "%":
  res = num1 % num2
else:
  print("Invalid operation")

print(f"Result: {res}")
```

```python
import random
check = input("Would you like to play? (yes/no): ")
while check == 'yes':
  pmove = input("Choose your action (r/p/s): ")
  cmove = random.choice(['r', 'p', 's'])
  print(f"Computer chooses: {cmove}")
  if pmove == 'r':
    if cmove == 'p':
      print("You lose!")
    if cmove == 's':
      print("You win!")
    if cmove == 'r':
      print("It's a tie.")
  if pmove == 'p':
    if cmove == 's':
      print("You lose!")
    if cmove == 'r':
      print("You win!")
    if cmove == 'p':
      print("It's a tie.")
  if pmove == 's':
    if cmove == 'r':
      print("You lose!")
    if cmove == 'p':
      print("You win!")
    if cmove == 's':
      print("It's a tie.")
  check = input("Would you like to play again? (yes/no): ")
```

```python
import random

play = input("Do you want to play a game (y/n)? ")

while play == 'y':
  player = input("Choose b/w rock, paper, scissors: ").lower()
  computer = random.choice(['rock', 'paper', 'scissors'])

  print(f"Player: {player}\nComputer: {computer}")

  if player == computer:
    print("Draw")
    elif (
      (player == 'rock'and computer == 'scissors') or
      (player == 'paper'and computer == 'rock') or
      (player == 'scissors'and computer == 'paper')
     ):
    else:
      print("Computer wins!")
  play = input("Do you want to play a game (y/n)? ")
```
